```python
from bs4 import BeautifulSoup
import requests
import json
import numpy as np
import pandas as pd
```

```python
text_class = 'text-infmlk'
label_class = 'label-infmlk'
```

```python
# sample web page
sample_web_page = 'https://www.melketabriz.com/p/31208#!'

# call get method to request that page
page = requests.get(sample_web_page)

# with the help of beautifulSoup and html parser create soup
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# Labels to check
labels = ["کد آگهی",
          "منطقه",
          "محله",
          "متراژ",
          "قیمت کل", 
          "طبقه",
          "تعداد اتاق",
          "سال ساخت",
          "نوع سند"
          ,"کل طبقات", 
          "تعداد واحد در طبقه",
          "موقعیت",
          "نوع ساخت",
          "زمان تحویل"
]

# Find all label and text elements
label_elements = soup.find_all(class_=label_class)
text_elements = soup.find_all(class_=text_class)

# Loop through the labels and match them with the text
for i in range(len(label_elements)):
    label_text = label_elements[i].find('div').get_text(strip=True)
    if label_text in labels:
        text = text_elements[i].find('div').get_text(strip=True)
        print(f"{label_text} : {text}")
        
```
